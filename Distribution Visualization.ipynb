### Distribution Visualization

Discrete and Continuous Variables

A discrete variable is a variable that can only take certain values. If you can count a set of items, then itâ€™s a discrete variable. e.g., The number of students in a class

A continuous variable takes on an any values. e.g., Height, length

In stats we represent distribution of discrete variables with PMF (Probability Mass Functions) and continuous variables with PDF (Probability Density Functions).

#### Visualizing a simple normal distribution with a mean of 0 and standard deviation of 1.

PDF of a normal distribution in formula form


### PDF normal distribution

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, scipy.stats.norm.pdf(x, mu, sigma), label='Normal Distribution')
plt.title('Normal Distribution with mean = 0 and std = 1')
plt.legend(fontsize='xx-large')
plt.show()
```
